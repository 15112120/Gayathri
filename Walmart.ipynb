```python
df['Store_Category']=0*len(df)
df
```

```python
list(df.columns)
```

```python
def categorise(s):
  if s['Weekly_Sales']<=500000:
    return "low"
  elif s['Weekly_Sales']>500000 and s['Weekly_Sales']<=1000000:
    return "medium"
  elif s['Weekly_Sales']>1000000 and s['Weekly_Sales']<= 1500000:
      return "High"
  else:
      return 'Super'
df['Store_Category']=df.apply(categorise,axis=1)
```

```python
df
```

```python
pv_store=pd.pivot_table(df,index=['Store'],aggfunc={'Weekly_Sales':'mean'})
pv_store
```
